```python
def get_pred(data, lag=2):
    d1 = data[FEATURES[:-lag]].apply(tuple, axis=1).to_frame().rename(columns={0: 'key'})
    d2 = data[FEATURES[lag:]].apply(tuple, axis=1).to_frame().rename(columns={0: 'key'})        
    d2['pred'] = data[FEATURES[lag-2]]
    d3 = d2[~d2.duplicated(['key'], keep=False)]
    return d1.merge(d3, how='left', on='key').pred.fillna(0)
```

```python
def get_all_pred(data, max_lag):
    target = pd.Series(index=data.index, data=np.zeros(data.shape[0]))
    for i in range(2, max_lag + 1):
        pred = get_pred(data, i)
        indices_to_be_changed = (target ==0 ) & (pred != 0)
        target[indices_to_be_changed] = pred[indices_to_be_changed]
    return target
```

```python
for max_lag in range(2, 30):
    predictions = get_all_pred(train, max_lag)
    non_zero_pred_indices = predictions != 0
    non_zero_pred = predictions[non_zero_pred_indices]
    target_corr_non_zero_pred = train.target[non_zero_pred_indices]
    print('##################')
    print('Max lag is  ', max_lag)
    print('Number of relevant predictions is ', len(non_zero_pred))
    print('Prediction score is ', np.sqrt(mean_squared_log_error(non_zero_pred, target_corr_non_zero_pred)))
```
